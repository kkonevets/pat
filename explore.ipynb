```python
with open('../data/gold.txt', 'r') as f:
    gold = f.read()
lines = gold.splitlines()
dist = pd.Series([len(l.split())-1 for l in lines], name='количество похожих патентов на заявку')

p = sns.distplot(dist,)

plt.show()
```
